```python
import random

class Grid:
    # Implementação do grid: execução
    def __init__(self, quicksand, mountain):
        quicksand_pos = []
    
        # Loop para gerar as posições aleatórias da areia movediça
        while len(quicksand_pos) < quicksand:
            position = random.randint(0,63)
            if position not in quicksand_pos:
                quicksand_pos.append(position)
        
        self.quicksand = quicksand_pos # lista das posições de areia movediça

        mountain_pos = []

        # Loop para gerar as posições aleatórias das montanhas
        while len(mountain_pos) < mountain:
            position = random.randint(0,63)
            if position not in quicksand_pos and position not in mountain_pos:
                mountain_pos.append(position)
                
        self.mountain = mountain_pos # lista das posições de montanhas

        # Geração aleatória da posição inicial, garantindo que não esteja em areia movediça ou montanhas
        start_pos = random.randint(0,63)
        while start_pos in quicksand_pos or start_pos in mountain_pos:
            start_pos = random.randint(0,63)
        
        self.start = start_pos # posição inicial

        # Geração aleatória da posição do objetivo, garantindo que não esteja em areia movediça, montanhas ou na posição inicial
        goal_pos = random.randint(0,63)
        while goal_pos in quicksand_pos or goal_pos in mountain_pos or goal_pos == start_pos:
            goal_pos = random.randint(0,63)
        
        self.goal = goal_pos # posição do objetivo
        
    # Função para visualização do grid
    def show_grid(self):        
        for i in range(8):
            print_row = ""
            for j in range(8):
                cell = i*8 + j
                if cell in self.mountain:
                    print_row += "m"
                elif cell in self.quicksand:
                    print_row += "q"
                elif cell == self.goal:
                    print_row += "g"
                elif cell == self.start:
                    print_row += "s"
                else:
                    print_row += "x"
                if j != 7:
                    print_row += " - "
            print(print_row)


class Agent:
    actions = {
        "up": -8,
        "down": 8,
        "right": 1,
        "left": -1
    }
    
    def __init__(self, grid):
        self.grid = grid
        self.position = grid.start
        self.reward = 0
    
    # Função com o conjunto de ações que o agente pode realizar
    def move(self, action):
        new_position = self.position + self.actions[action]
        
        # Verifica se a nova posição é válida e não é uma montanha
        if new_position < 64 and new_position >= 0 and new_position not in self.grid.mountain:
            self.position = new_position
            return True
        else:
            return False
            
    # Função com as recompensas
    def reward_function(self):
        # Atribui recompensas com base na posição do agente
        if self.position == self.grid.goal:
            self.reward += 10
            return True
        elif self.position in self.grid.quicksand:
            self.reward -= 10
            return True
        return False
    
    # Função com a Lógica do Agente
    def game_over(self):
        # Verifica se o jogo acabou (o agente alcançou o objetivo ou caiu em areia movediça)
        return self.position == self.grid.goal or self.position in self.grid.quicksand


grid = Grid(quicksand=4, mountain=8)
agent = Agent(grid)

while not agent.game_over():
    action = random.choice(list(agent.actions.keys()))
    if agent.move(action):
        agent.reward_function()

grid.show_grid()
print("Reward:", agent.reward)
```
